### Absorption Rate

#### Data

```python
[ masses_eV, abs_rates_1 ] = data.get_absorption_rates(g2 = 10**(-26),
                                                      width = [0.2, 0.1, 1000])

[ masses_eV, abs_rates_2 ] = data.get_absorption_rates(g2 = 10**(-26),
                                                      width = [1, 1, 1])
```

### $g_{aee}$ Constraint

#### Data

```python
[ masses_eV, g_constraint_1 ] = data.get_abs_g_constraint(width = [0.2, 0.1, 1000])

[ masses_eV, g_constraint_2 ] = data.get_abs_g_constraint(width = [1, 1, 1])
```

#### Plot

```python
save_fig = True

fig, axes = plt.subplots(nrows=1, ncols=1, 
                             figsize=(7*1.1, 7))

log_mX_min = 0
log_mX_max = 3

log_events_min = -14
log_events_max = -9

set_custom_tick_options(axes)
set_custom_axes(axes, 'x', log_mX_min, log_mX_max,
               ax_type = 'log', 
               label = r'$m_\phi \, [\mathrm{eV}]$')
set_custom_axes(axes, 'y', log_events_min, log_events_max,
               ax_type = 'log', 
               label = r'$g_{aee}$', 
               show_first = False)

axes.plot(
    np.log10(masses_eV), 
    np.log10(g_constraint_1),
)

axes.plot(
    np.log10(masses_eV), 
    np.log10(g_constraint_2),
)

fig.tight_layout()
    
if save_fig:
    plt.savefig('./output/Si_gaee_constraint_2x2x2.pdf', 
                bbox_inches='tight', pad_inches = 0.075)

plt.show()
```
