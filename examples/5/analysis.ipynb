### Binned Scatter Rate (with example 2 results)

#### Data

```python
data_ex1 = EXDMData(filename = '../1/output/EXDM_out_example_1.hdf5')
[ E_bin_LHS_ex1, binned_scatter_rate_E_ex1 ] = data_ex1.get_binned_scatter_rate_E(mass_MeV = 10**3, 
                                                                      med_FF = 0., 
                                                                      sigma_cm2 = 10**(-40),
                                                                      E_bin_width = 10.)

data_ex2 = EXDMData(filename = '../2/output/EXDM_out_example_2.hdf5')
[ E_bin_LHS_ex2, binned_scatter_rate_E_ex2 ] = data_ex2.get_binned_scatter_rate_E(mass_MeV = 10**3, 
                                                                      med_FF = 0., 
                                                                      sigma_cm2 = 10**(-40),
                                                                      E_bin_width = 10.)

data_ex3 = EXDMData(filename = '../3/output/EXDM_out_example_3.hdf5')
[ E_bin_LHS_ex3, binned_scatter_rate_E_ex3 ] = data_ex3.get_binned_scatter_rate_E(mass_MeV = 10**3, 
                                                                      med_FF = 0., 
                                                                      sigma_cm2 = 10**(-40),
                                                                      E_bin_width = 10.)

data_ex4 = EXDMData(filename = '../4/output/EXDM_out_example_4.hdf5')
[ E_bin_LHS_ex4, binned_scatter_rate_E_ex4 ] = data_ex4.get_binned_scatter_rate_E(mass_MeV = 10**3, 
                                                                      med_FF = 0., 
                                                                      sigma_cm2 = 10**(-40),
                                                                      E_bin_width = 10.)
```
